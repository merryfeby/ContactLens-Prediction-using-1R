```python
import pandas as pd
from mlxtend.classifier import OneRClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('ContactLens.csv')
df.head()
```

```python
le = LabelEncoder()
for col in df.columns:
  df[col] = le.fit_transform(df[col])
```

```python
x = df.drop('ContactLens', axis=1)
x
```

```python
y = df['ContactLens']
y
```

```python

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Train OneR classifier
model = OneRClassifier()
model.fit(x_train.values, y_train) 

#Predict the response for test dataset
y_pred = model.predict(x_test.values) 
y_pred

# x_train.values and x_test.values: By using .values,extract the underlying NumPy array from the Pandas DataFrames x_train and x_test. 
# This provides the classifier with the expected array-like input.
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```
